```python
import numpy as np
import pandas as pd

df_data = pd.read_csv("data_analytics.csv")
df_data.head()
```

```python
#delete not informative columns:

del df_data['App Apple ID']
del df_data['Subscription Name']
del df_data['Standard Subscription Duration']
del df_data['Promotional Offer Name']
del df_data['Marketing Opt-In Duration']
del df_data['Promotional Offer ID']
del df_data['Subscriber ID Reset']
del df_data['Purchase Date']
del df_data['Refund']
del df_data['Units']

df_data
```

```python
subscribers = df_data['Subscriber ID'].tolist()

#unique IDs:
subscribers = set(subscribers)
subscribers = list(subscribers)

df_data.rename(columns={'Subscriber ID': 'Subscriber_ID'}, inplace=True)
df_data
```

```python
LTVs = []
price = 9.99
apple_store_tax = 0.3

for ID in subscribers:
    weeks = len(df_data[df_data.Subscriber_ID == ID]['Subscriber_ID'])
    LTV = (weeks-1)*(1-apple_store_tax)*price
    LTVs.append(LTV)

LTVs
```

```python
LTV_df = pd.DataFrame({'Subscriber ID': subscribers, 'LTV': LTVs})
LTV_df
```

```python
LTV_df.to_csv("LTV.csv", index=False)
```
